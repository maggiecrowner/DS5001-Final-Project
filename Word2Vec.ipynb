```python
import gensim
gensim.__version__
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio
from gensim.models import word2vec
from sklearn.manifold import TSNE as tsne
```

```python
!wget -O CORPUS.csv "https://virginia.box.com/shared/static/ijkqovrdgvrmctsdqobymig2p98q9x0m.csv"
!wget -O LIB.csv "https://virginia.box.com/shared/static/fhzudg34je9xls5bfcbi4xdnaiek74rj.csv"
```

```python
CORPUS = pd.read_csv('CORPUS.csv', delimiter='|', index_col=[0,1,2,3])
LIB = pd.read_csv('LIB.csv', delimiter='|', index_col=0)
```

##VOCAB_W2V

```python
BAG = ['Artist', 'Album', 'Title']
docs = CORPUS.groupby(BAG).term_str.apply(list).tolist()
```

```python
for i in range(5):
    print(f"Doc {i}:", docs[i])
```

```python
w2v_params = dict(
    window = 2,
    vector_size = 256,
    min_count = 50,
    workers = 1,
    seed = 42
)
```

```python
model = word2vec.Word2Vec(docs, **w2v_params)
```

```python
VOCAB_W2V = pd.DataFrame(model.wv.vectors, index=model.wv.index_to_key)
VOCAB_W2V.index.name = 'term_str'
VOCAB_W2V.head()
```

```python
VOCAB_W2V.to_csv("/content/DS5001-Final-Project/VOCAB_W2V.csv", sep="|", index=True)
```

## Word2vec tSNE Plot

```python
tsne_params = dict(
    perplexity=40,
    n_components=2,
    init='pca',
    max_iter=2500,
    random_state=43
)

tsne_engine = tsne(**tsne_params)

TSNE = pd.DataFrame(
    tsne_engine.fit_transform(VOCAB_W2V),
    columns=['x','y'],
    index=VOCAB_W2V.index)

TSNE.head()
```
