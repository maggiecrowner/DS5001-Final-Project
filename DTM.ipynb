```python
import pandas as pd
BOW = pd.read_csv('BOW.csv', delimiter='|', index_col=["Artist", "Album", "Title", "term_str"])
```

```python
DTM = BOW.n.unstack(fill_value=0)
DTM.head()
```

```python
DTM.to_csv('/content/DS5001-Final-Project/DTM.csv', sep='|', index=True)
```
